```python
#function for computing similarity score for each pair of text
@cache
def compute_similarity_score(dataframe , len_of_dataframe):
  # encode sentences to get their embeddings
  embedding1 = model.encode(sentence1, convert_to_tensor=True)
  embedding2 = model.encode(sentence2, convert_to_tensor=True)
# compute similarity scores of two embeddings
  cosine_scores = util.pytorch_cos_sim(embedding1, embedding2)
  return cosine_scores.item()

score = compute_similarity_score(sentence1,sentence2)
score
```

```python
#compute similarity score for every pair
similarity_scores = []
for i in range(len(df)):
  similarity_scores.append(compute_similarity_score(df['text1'][i],df['text2'][i]))

similarity_score[:10]
```

```python
#load progess
df = pd.read_csv("/content/drive/MyDrive/Precily AI/text_similarity2.csv")
df= df[['text1','text2','similarity','result']]
df.head()
```

```python
#now extract embeddings
text1_embedding, text2_embedding = [],[]
df_len = len(df)
```

```python

embedding1 = model.encode(df['text1'][:], convert_to_tensor=True)
embedding2 = model.encode(df['text2'][:], convert_to_tensor=True)
```

```python
embedding1[:5]
```

```python
#now concatinate embedding1 and embedding2 for training as X numpy array
embedding1_cpu = embedding1.cpu().numpy()
embedding2_cpu = embedding2.cpu().numpy()

X = np.concatenate([embedding1_cpu, embedding2_cpu], axis=1)
X[:5]
```

```python
# target variable
Y = np.asarray(df['similarity'])
Y[:5]
```

```python
#import tensorflow lib for modelling
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
#from tensorflow.keras.metrics import mean_absolute_percentage_error
from sklearn.model_selection import train_test_split
```

```python
#splitting the dataset
trainx , testx, trainy, testy = train_test_split(X,Y,test_size=0.2);
```

```python
#calculating max_length of sequences in to pass it as parameter in Dense layer: input_dim
seq_len = [len(i) for i in X ]
max_sequence_length = max(seq_len)
```

```python

#tensorflow regression model: train-test
mod = Sequential()
mod.add(Dense(64, input_dim= max_sequence_length, activation = 'relu' ))
mod.add(Dense(32,activation='relu'))
mod.add(Dense(1))
mod.compile(loss='mse',metrics = [tf.keras.metrics.MeanAbsolutePercentageError()], optimizer='adam')
mod.summary()
```

```python
mod.fit(trainx,trainy, epochs = 10)
```

```python
#predict test set
pred2 = mod.predict(testx)
for i in range(20):
  tf.print(pred2[i])
```

```python
testy[:20]
```

```python
#evaluate model
mod.evaluate(testx,testy)
#eval shows mse as large but if i extract labels 1(similar) and  0(not similar) we get:
```

```python
pred_label,test_label = [] ,[]
for i in range(20):
  if pred2[i]>.5:
    pred_label.append(1)
  else:
    pred_label.append(0)

  if testy[i]>.5:
    test_label.append(1)
  else:
    test_label.append(0)

print(pred_label,test_label)

#as you can see after extracting labels we have highly accurate regression model
```

```python
#tensorflow regression model all data
model = Sequential()
model.add(Dense(64, input_dim= max_sequence_length, activation = 'relu' ))
model.add(Dense(32, activation = 'relu' ))
model.add(Dense(1))

model.compile(loss='mse',metrics = [tf.keras.metrics.MeanAbsolutePercentageError()], optimizer='adam')
model.summary()
model.fit(X,Y, epochs=20)
```
